```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Use California housing dataset as alternative
housing_data = fetch_california_housing()
x = housing_data.data
y = housing_data.target

# Create DataFrame with feature names
data = pd.DataFrame(x, columns=housing_data.feature_names)
data["SalePrice"] = y  # Adding target variable as SalePrice column
data.head()
```

```python
#to find highly correlated features
cor_target = abs(cor["SalePrice"]) #absolute value of corelation
relevant_features = cor_target[cor_target > 0.2] #highly correlayed features
names = [index for index, value in relevant_features.items()] #getting names of features

names.remove('SalePrice') #removing target feature
print(names)
print(len(names))
```

```python
#MODEL BUILDING
```

```python
from sklearn.model_selection import train_test_split

x = data.drop("SalePrice", axis=1)
y= data["SalePrice"]
x_train,x_test, y_train, y_test =train_test_split(x,y, test_size =0.2, random_state=42) #random state is to make sure that data is not change after running
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)

print("actual value of house : ",y_test[0])
print("model predicted value : ",predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,predictions)
print(mse)
```
